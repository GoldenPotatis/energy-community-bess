```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Create figure with two subplots (for peak power and costs)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# Data preparation
months = ['January', 'February', 'March', 'April', 'Average']

# Peak power data (kW)
before_peaks = [105.92, 139.58, 146.94, 170.48, 140.73]
after_system_peaks = [84.79, 98.06, 97.15, 98.31, 94.58]
after_bess_peaks = [63.79, 75.23, 78.61, 67.30, 71.23]

# Cost data (SEK)
before_costs = [7732.16, 10189.34, 10726.62, 12445.02, 10273.29]
after_system_costs = [6189.67, 7158.02, 7091.95, 7176.63, 6904.07]
after_bess_costs = [4656.67, 5738.53, 5738.53, 4912.90, 5199.98]

# Set width of bars
barWidth = 0.25
r1 = np.arange(len(months))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Create grouped bar chart for peak power
ax1.bar(r1, before_peaks, width=barWidth, edgecolor='grey', label='Before Interventions', color='#1f77b4')
ax1.bar(r2, after_system_peaks, width=barWidth, edgecolor='grey', label='After System Improvements', color='#ff7f0e')
ax1.bar(r3, after_bess_peaks, width=barWidth, edgecolor='grey', label='After BESS Installation', color='#2ca02c')

# Create grouped bar chart for costs
ax2.bar(r1, before_costs, width=barWidth, edgecolor='grey', label='Before Interventions', color='#1f77b4')
ax2.bar(r2, after_system_costs, width=barWidth, edgecolor='grey', label='After System Improvements', color='#ff7f0e')
ax2.bar(r3, after_bess_costs, width=barWidth, edgecolor='grey', label='After BESS Installation', color='#2ca02c')

# Add labels and titles
ax1.set_title('Maximum Monthly Peak Power', fontsize=14, fontweight='bold')
ax1.set_ylabel('Peak Power (kW)', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax1.legend(loc='upper left')

ax2.set_title('Monthly Power Costs', fontsize=14, fontweight='bold')
ax2.set_ylabel('Cost (SEK)', fontsize=12)
ax2.set_xlabel('Month', fontsize=12)
ax2.grid(axis='y', linestyle='--', alpha=0.7)
ax2.legend(loc='upper left')

# Adjusting x-axis
plt.xticks([r + barWidth for r in range(len(months))], months, rotation=45)

# Add percentage reduction annotations for the average values
avg_idx = 4  # Index of "Average" in the data
peak_red_sys = round((before_peaks[avg_idx] - after_system_peaks[avg_idx])/before_peaks[avg_idx]*100, 1)
peak_red_bess = round((after_system_peaks[avg_idx] - after_bess_peaks[avg_idx])/after_system_peaks[avg_idx]*100, 1)
cost_red_sys = round((before_costs[avg_idx] - after_system_costs[avg_idx])/before_costs[avg_idx]*100, 1)
cost_red_bess = round((after_system_costs[avg_idx] - after_bess_costs[avg_idx])/after_system_costs[avg_idx]*100, 1)

ax1.annotate(f'-{peak_red_sys}%', xy=(r2[avg_idx], after_system_peaks[avg_idx]), 
             xytext=(0, 10), textcoords='offset points', ha='center', fontweight='bold')
ax1.annotate(f'-{peak_red_bess}%', xy=(r3[avg_idx], after_bess_peaks[avg_idx]), 
             xytext=(0, 10), textcoords='offset points', ha='center', fontweight='bold')

ax2.annotate(f'-{cost_red_sys}%', xy=(r2[avg_idx], after_system_costs[avg_idx]), 
             xytext=(0, 10), textcoords='offset points', ha='center', fontweight='bold')
ax2.annotate(f'-{cost_red_bess}%', xy=(r3[avg_idx], after_bess_costs[avg_idx]), 
             xytext=(0, 10), textcoords='offset points', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('peak_power_cost_reduction.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# Create figure with two subplots (horizontally arranged)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Data preparation
months = ['January', 'February', 'March', 'April', 'Average']

# Peak power data (kW)
before_peaks = [105.92, 139.58, 146.94, 170.48, 140.73]
after_system_peaks = [84.79, 98.06, 97.15, 98.31, 94.58]
after_bess_peaks = [63.79, 75.23, 78.61, 67.30, 71.23]

# Cost data (SEK)
before_costs = [7732.16, 10189.34, 10726.62, 12445.02, 10273.29]
after_system_costs = [6189.67, 7158.02, 7091.95, 7176.63, 6904.07]
after_bess_costs = [4656.67, 5738.53, 5738.53, 4912.90, 5199.98]

# Define Morandi colors
morandi_colors = ['#a99f97', '#e1d7cd', '#8a7f76']

# Set width of bars
barWidth = 0.25
r1 = np.arange(len(months))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

# Create grouped bar chart for peak power
bars1 = ax1.bar(r1, before_peaks, width=barWidth, edgecolor='grey', label='Before Interventions', color=morandi_colors[0])
bars2 = ax1.bar(r2, after_system_peaks, width=barWidth, edgecolor='grey', label='After System Improvements', color=morandi_colors[1])
bars3 = ax1.bar(r3, after_bess_peaks, width=barWidth, edgecolor='grey', label='After BESS Installation', color=morandi_colors[2])

# Create grouped bar chart for costs
bars4 = ax2.bar(r1, before_costs, width=barWidth, edgecolor='grey', label='Before Interventions', color=morandi_colors[0])
bars5 = ax2.bar(r2, after_system_costs, width=barWidth, edgecolor='grey', label='After System Improvements', color=morandi_colors[1])
bars6 = ax2.bar(r3, after_bess_costs, width=barWidth, edgecolor='grey', label='After BESS Installation', color=morandi_colors[2])

# Add labels and titles
ax1.set_title('Maximum Monthly Peak Power', fontsize=14, fontweight='bold')
ax1.set_ylabel('Peak Power (kW)', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax1.legend(loc='upper left')

ax2.set_title('Monthly Power Costs', fontsize=14, fontweight='bold')
ax2.set_ylabel('Cost (SEK)', fontsize=12)
ax2.grid(axis='y', linestyle='--', alpha=0.7)
ax2.legend(loc='upper left')

# Adjusting x-axis for both subplots
ax1.set_xticks([r + barWidth for r in range(len(months))])
ax1.set_xticklabels(months, rotation=45)
ax2.set_xticks([r + barWidth for r in range(len(months))])
ax2.set_xticklabels(months, rotation=45)

# Function to add value labels on bars
def add_value_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax = bar.axes
        ax.annotate(f'{int(round(height, 0))}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=9)

# Add value labels to all bars
add_value_labels(bars1)
add_value_labels(bars2)
add_value_labels(bars3)
add_value_labels(bars4)
add_value_labels(bars5)
add_value_labels(bars6)

# Add percentage reduction annotations for the average values
avg_idx = 4  # Index of "Average" in the data
peak_red_sys = round((before_peaks[avg_idx] - after_system_peaks[avg_idx])/before_peaks[avg_idx]*100, 0)
peak_red_bess = round((after_system_peaks[avg_idx] - after_bess_peaks[avg_idx])/after_system_peaks[avg_idx]*100, 0)
cost_red_sys = round((before_costs[avg_idx] - after_system_costs[avg_idx])/before_costs[avg_idx]*100, 0)
cost_red_bess = round((after_system_costs[avg_idx] - after_bess_costs[avg_idx])/after_system_costs[avg_idx]*100, 0)

ax1.annotate(f'-{int(peak_red_sys)}%', xy=(r2[avg_idx], after_system_peaks[avg_idx]), 
             xytext=(0, -20), textcoords='offset points', ha='center', fontweight='bold')
ax1.annotate(f'-{int(peak_red_bess)}%', xy=(r3[avg_idx], after_bess_peaks[avg_idx]), 
             xytext=(0, -20), textcoords='offset points', ha='center', fontweight='bold')

ax2.annotate(f'-{int(cost_red_sys)}%', xy=(r2[avg_idx], after_system_costs[avg_idx]), 
             xytext=(0, -20), textcoords='offset points', ha='center', fontweight='bold')
ax2.annotate(f'-{int(cost_red_bess)}%', xy=(r3[avg_idx], after_bess_costs[avg_idx]), 
             xytext=(0, -20), textcoords='offset points', ha='center', fontweight='bold')

plt.tight_layout()
plt.savefig('peak_power_cost_reduction.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
scenarios = ['1.a', '1.b', '1.c', '2.a', '2.b', '2.c', '3.a', '3.b', '3.c']
peak_shaving = ['10% (6 kW)', '10% (6 kW)', '10% (6 kW)', 
                '40% (24 kW)', '40% (24 kW)', '40% (24 kW)',
                '80% (48 kW)', '80% (48 kW)', '80% (48 kW)']
grid_comp = ['0.05 SEK/kW', '0.25 SEK/kW', '0.45 SEK/kW',
             '0.05 SEK/kW', '0.25 SEK/kW', '0.45 SEK/kW',
             '0.05 SEK/kW', '0.25 SEK/kW', '0.45 SEK/kW']

# Revenue and payback period data
yearly_revenue = [23512, 96536, 169560, 38720, 109504, 180288, 56524, 114428, 172332]
yearly_revenue_k = [y/1000 for y in yearly_revenue]  # Convert to kSEK
payback_period = [25.5, 6.2, 3.5, 15.5, 5.5, 3.3, 10.6, 5.2, 3.5]  # Rounded to 1 decimal

# Set up the figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Create indices for grouped bars
indices = np.arange(len(scenarios))
bar_width = 0.6

# Use the same color for each scenario group
colors = []
for i in range(len(scenarios)):
    if i < 3:  # Scenario 1 (a,b,c)
        colors.append('#4682B4')  # Blue for 10% peak shaving
    elif i < 6:  # Scenario 2 (a,b,c)
        colors.append('#2E8B57')  # Green for 40% peak shaving
    else:  # Scenario 3 (a,b,c)
        colors.append('#CD5C5C')  # Red for 80% peak shaving

# Plot yearly revenue
bars1 = ax1.bar(indices, yearly_revenue_k, bar_width, color=colors, linewidth=0.5)
ax1.set_title('Yearly Revenue by Scenario', fontsize=14, fontweight='bold')
ax1.set_ylabel('Yearly Revenue (kSEK)', fontsize=12)
ax1.set_xticks(indices)
ax1.set_xticklabels(scenarios)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Plot payback period
bars2 = ax2.bar(indices, payback_period, bar_width, color=colors, linewidth=0.5)
ax2.set_title('Payback Period by Scenario', fontsize=14, fontweight='bold')
ax2.set_ylabel('Payback Period (years)', fontsize=12)
ax2.set_xticks(indices)
ax2.set_xticklabels(scenarios)
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# Add value labels on top of bars
def add_value_labels(bars, decimals=1):
    for bar in bars:
        height = bar.get_height()
        ax = bar.axes
        format_str = '{:.1f}' if decimals == 1 else '{:.0f}'
        ax.annotate(format_str.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=9)

# Add value labels
add_value_labels(bars1, decimals=1)  # 1 decimal for kSEK
add_value_labels(bars2, decimals=1)  # 1 decimal for years

# Add legend only to the right panel (payback period)
legend_elements = [
    plt.Rectangle((0,0), 1, 1, fc='#4682B4', edgecolor='grey', label='10% Peak Shaving'),
    plt.Rectangle((0,0), 1, 1, fc='#2E8B57', edgecolor='grey', label='40% Peak Shaving'),
    plt.Rectangle((0,0), 1, 1, fc='#CD5C5C', edgecolor='grey', label='80% Peak Shaving')
]

# Add frameless legend to the right panel
leg = ax2.legend(handles=legend_elements, loc='upper right', fontsize=11)
leg.get_frame().set_linewidth(0.0)  # Remove frame

# Highlight the best performing scenario (2.c for revenue, 2.c for payback)
bars1[5].set_edgecolor('black')  # Scenario 2.c for revenue
bars1[5].set_linewidth(2)
bars2[5].set_edgecolor('black')  # Scenario 2.c for payback
bars2[5].set_linewidth(2)

# Add grid service compensation annotation below the legend in right panel
ax2.text(0.75, 0.7, "Grid Service Compensation:", transform=ax2.transAxes, 
         ha='center', fontsize=11, fontweight='bold')
ax2.text(0.75, 0.48, "Scenarios .a: 0.05 SEK/kW\n\nScenarios .b: 0.25 SEK/kW\n\nScenarios .c: 0.45 SEK/kW", 
         transform=ax2.transAxes, ha='center', fontsize=11)

plt.tight_layout(rect=[0, 0.08, 1, 0.95])  # Allow space for bottom annotation
plt.savefig('scenario_modeling_results.png', dpi=300, bbox_inches='tight')
plt.show()
```
